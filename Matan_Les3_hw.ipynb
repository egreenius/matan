1. Последовательность $\{a_n\}_{n=1}^\infty = 2^n - n$ является 
возрастающей и неограниченной. 5 й член последовательности = 27

*   возрастающей так как разница n+1 и n -го члена  > 0:  $2^{n+1} - (n + 1) - (2^n - n) = 2^{n+1} - n - 1 - 2^n + n = 2^{n+1} - 1 - 2^n = 2^n(2-1) -1 = 2^n - 1 > 0, \implies \forall n \in N : a_{n+1} > a_n$
*   неограниченной сверху так как $a_{n+1} - a_n = 2^n -1 \to \infty$ с ростом n
* пятый член последовательности $a_5 = 2^5 - 5 = 27$

2. Последовательность  $\{b_n\}_{n=2}^\infty = \frac{1}{1-n}$ является возрастающей и ограниченной сверху 0. 5 - й член последовательности = $-\frac{1}{5}$

*   возрастающей так как последовательность состоит из отрицательных членов и отношение n+1 и n -го члена < 1:
$\frac{b_{n+1}}{b_n} = \frac{\frac{1}{1-(n+1)}}{\frac{1}{1-n}} = \frac{1}{-n} * \frac{1-n}{1} = \frac{n-1}{n} = 1 - \frac{1}{n} < 1 \implies \forall n \in [2; \infty) : b_{n+1} < b_n$
*   ограниченной: $-1 \le b_n < 0$ так как минимальным является первый член последовательности (-1)( при n = 2), а при $n \to \infty$, отношение следующего к предыдущему стремится к 1, а значит сама последовательность к 0.
*   пятый член последовательности $b_6 = \frac{1}{1-6} = -\frac{1}{5}$

3. Последовательность $\{c_n\}_{n=1}^\infty = -1^n + \sqrt{2n}$ не является мнотонной и не является ограниченной сверху. Пятый член последовательности равен $-1 + \sqrt10$

*   не является монотонной, так как четные элементы больше предыдущего и последующего нечетного, при этом отдельно четные и отдельно нечетные члены возрастают с ростом n. Действительно, посмотрим чему равна разность n-го члена $c_n$ и n+1 - го члена $c_{n+1}$: $-1^n + \sqrt{2n} - (-1^{n+1} + \sqrt{2(n+1)}) = -1^n - (-1^{n+1}) +\sqrt{2n} - \sqrt{2(n+1)} = -1^n - (-1^{n+1}) + \sqrt2(\sqrt{n} - \sqrt{n+1})$. Далее, рассмотрим два случая, когда n -  четное, и когда n - нечетное.

$
\begin{cases}
-1 - 1 + \sqrt2(\sqrt{n} - \sqrt{n+1}) & \quad \text{if } n \text{ is odd}\\
1+1 + \sqrt2(\sqrt{n} - \sqrt{n+1}) & \quad \text{if } n \text{ is even}
\end{cases}
$

или 

$
\begin{cases}
-2 + \sqrt2(\sqrt{n} - \sqrt{n+1}) & \quad \text{if } n \text{ is odd}\\
2 + \sqrt2(\sqrt{n} - \sqrt{n+1}) & \quad \text{if } n \text{ is even}
\end{cases}
$

выражение в скобках $(\sqrt{n} - \sqrt{n+1})$ всегда отрицательное, тогда разность $c_n - c_{n+1}$ для нечетных n всегда отрицательна, то есть $c_n \lt c_{n+1}$.
Для четных n максмальное значение разницы при n = 1 и равно примерно -0,4. Таким образом, для четных n разность членов $c_n - c_{n+1}$ будет всегда положительна, то есть $c_n \gt c_{n+1}$.

Так как четные и нечетные члены в последовательности чередуются с ростом n, то отношение предыдущего к последующему также постоянно чередуется с меньше на больше, что говорит о том, что последовательность не монотонна
*   не ограничена сверху, так как не существует числа K, для которого верно высказывание $\forall n \to \infty : |c_n| \le K$, так как нечетные элементы и четные элементы возрастают с ростом n. 
*   пятый член последовательности $c_5 = -1^5 + \sqrt{2*5} = -1 + \sqrt10$

4. Последовательность  $\{d_n\}_{n=1}^\infty = (-1)^{2n} + \frac{1}{n^2}$ является убывающей и ограниченной. Пятый член последовательности равен $1\frac{1}{25}$


*   найдем разность n - го и (n+1)-го члена последовательности $d_n - d_{n+1}= (-1)^{2n} + \frac{1}{n^2} - (-1)^{2(n+1)}- \frac{1}{(n+1)^2} = \frac{1}{n^2} - \frac{1}{(n+1)^2} \gt 0$ при $n \in [1; \infty): \implies d_n \gt d_{n+1}$  (следующий член меньше предыдущего), то есть последовательность убывающая. 
*   последовательность ограничена сверху при n = 1 и снизу при $n \to \infty$: $d_1 = 2$ и $d_{n\to\infty} = 1$

### Задание 3

*На языке Python предложить алгоритм вычисляющий численно предел с точностью $\epsilon = 10^{-7}$

$\lim\limits_{n \to \infty} \frac{n}{\sqrt[n]{n!}}$

```python
import math, sys

# задаем точность
eps = 0.0000001
print(f'Точность вычисление предела последовательности = {eps}')
# вычисляем первые два члена
n = 1
x1 = 1
n = n + 1
xn = n*math.pow(n*(n-1), - (1/n))
# в цикле вычисляем следующий член последовательности до тех пора пока разница с предыдущим
# не будет меньше или равна заданной точности
# sys.maxsize - наиболее близкая реализация в Python понятия максимального "натурального" целого
while (abs(xn - x1) > eps) and (n < sys.maxsize):
  n = n +1
  x1 = xn
  y = 1
  # вычисляем значение факторила в цикле сразу возводя каждый элемент в нужную степень
  for i in range(1, n+1):
    y_t = math.pow(i, 1/n)
    y = y * y_t
  xn = n/y

if n == sys.maxsize:
  print('Точность не достигнута')
else:
  print(f'Вычисленный предел последовательности с точностью {eps} равен {xn}, при n = {n}')
```
