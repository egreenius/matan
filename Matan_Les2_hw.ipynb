## Задание 1

Даны три множества a,b и с. Необходимо выполнить все изученные виды бинарных операций над всеми комбинациями множеств.

Пусть заданы множества:
a = {1; 20; 3; q}
b = {20; 6; 3; hello}
c = {3; world; !}

1. Объединение:
$a∪b∪c =$ {1; 20; 3; q; 6; hello; world; !}

2. Пересечение:
$a∩b∩c =$ {3}

3. Разность:
a\b\c = {1; q}

4. Симметрическая разность:
$aΔbΔc = $ {1; q; 6; hello; 3; world; !}

## Задание 2

*Выполнить задание 1 на языке Python

```python
a = {1, 20, 3, "q"} 
b = {20, 6, 3, "hello"} 
c = {3, "world", "!"}

```

```python
# Объединение
print(a|b|c)
```

```python
# Пересечение
print(a&b&c)
```

```python
# Разность
print(a - b - c)
```

```python
# Симметрическая разность
print(a^b^c)
```
